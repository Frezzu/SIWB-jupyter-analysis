# Hipotezy
## 1. Wraz z wiekiem wzrasta chęć zostanie menedżerem

```python
data.hist(column='Age', by='MgrWant', bins=15, sharex=True, sharey=True, figsize=(10, 10))
```

Analizując powyższe histogramy możemy dojść do wniosku, że wraz z wiekiem zainteresowanie posadą menedżera spada. Jest ono zdecydowanie wysokie między 20 a 30 rokiem życia - czyli na początku kariery. Z wykresów widzimy również, że im osoba starsza tym częsciej zaznaczała odpowiedź "No", co świadczy o spadku zainteresowania posadą menedżerską.

## 2. Programiści piszący testy zarabiają więcej.

```python
data.boxplot(column=['ConvertedComp'], by='UnitTests', figsize=(10, 10), showfliers=False, vert=False)
```

Analizując powyższy wkres zauważamy, że największą mediane zarobku zauważamy dla odpowiedzi "Yes, it's part of our process", co jednoznacznie świadczy o poprwaności tej hipotezy. Na wykresie widać, że jeśli testy nie są pisane w to najczęstsza płaca pracowników jest niższa. Dodatkowo zauważmy fakt, że tam gdzie testy są pisane, najniższa płaca jest o wiele wyższa niż w innych przypadkach (pomijając wartości odstające).
